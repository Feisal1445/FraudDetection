```python
import pandas as pd
df = pd.read_csv('creditcard.csv')
df
```

```python
# Import the RobustScaler class from the sklearn.preprocessing library
from sklearn.preprocessing import RobustScaler

# Create a copy of the original DataFrame and assign it to the new_df variable
new_df = df.copy()

# Apply the RobustScaler transformation to the 'Amount' column of the new_df DataFrame
# The fit_transform method is used to scale the data and transform it into a new range
# The to_numpy method is used to convert the 'Amount' column to a numpy array
# The reshape method is used to reshape the numpy array into a 2-dimensional array with a single column
# The resulting scaled and transformed data is assigned back to the 'Amount' column of the new_df DataFrame
new_df['Amount'] = RobustScaler().fit_transform(new_df['Amount'].to_numpy().reshape(-1, 1))

# Extract the 'Time' column from the new_df DataFrame and assign it to the time variable
time = new_df['Time']

# Scale the 'Time' column using min-max scaling
# The min-max scaling method scales the data into the range of 0 to 1
# The resulting scaled data is assigned back to the 'Time' column of the new_df DataFrame
new_df['Time'] = (time - time.min()) / (time.max() - time.min())
```

```python
# Shuffle the rows of the new_df DataFrame using a random_state of 1
# The frac parameter sets the fraction of rows to be randomly sampled, in this case, all rows are included
new_df = new_df.sample(frac=1, random_state=1)

# Display the shuffled DataFrame
new_df
```

```python
# Split the DataFrame into train, test, and val subsets
train, test, val = new_df[:240000], new_df[240000:260000], new_df[260000:]

# Count the instances of 'Class' column in each subset
train['Class'].value_counts(), test['Class'].value_counts(), val['Class'].value_counts()
```

```python
# Convert the train, test, and val DataFrames into numpy arrays
train_np, test_np, val_np = train.to_numpy(), test.to_numpy(), val.to_numpy()

# Print the shapes of the resulting numpy arrays
train_np.shape, test_np.shape, val_np.shape
```

```python
# Split the train, test, and val numpy arrays into input (x) and output (y) variables
x_train, y_train = train_np[:, :-1], train_np[:, -1]
x_test, y_test = test_np[:, :-1], test_np[:, -1]
x_val, y_val = val_np[:, :-1], val_np[:, -1]

# Print the shapes of the input and output variables for each subset
x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

```python
# Import the LogisticRegression class from sklearn.linear_model
from sklearn.linear_model import LogisticRegression

# Create an instance of the LogisticRegression class
logistic = LogisticRegression()

# Fit the logistic regression model to the training data
logistic.fit(x_train, y_train)

# Calculate the accuracy score of the logistic regression model on the training data
logistic.score(x_train, y_train)
```

```python
# Import the classification_report function from sklearn.metrics
from sklearn.metrics import classification_report

# Print the classification report of the logistic regression model on the validation data
print(classification_report(y_val, logistic.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Import the RandomForestClassifier class from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the RandomForestClassifier class with max_depth of 2 and n_jobs set to -1
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)

# Fit the random forest classifier to the training data
rf.fit(x_train, y_train)

# Print the classification report of the random forest classifier on the validation data
print(classification_report(y_val, rf.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Import the GradientBoostingClassifier class from sklearn.ensemble
from sklearn.ensemble import GradientBoostingClassifier

# Create an instance of the GradientBoostingClassifier class with n_estimators=50, learning_rate=1.0, max_depth=1, and random_state=0
gbc = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)

# Fit the gradient boosting classifier to the training data
gbc.fit(x_train, y_train)

# Print the classification report of the gradient boosting classifier on the validation data
print(classification_report(y_val, gbc.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Import the LinearSVC class from sklearn.svm
from sklearn.svm import LinearSVC

# Create an instance of the LinearSVC class with class_weight set to 'balanced'
svc = LinearSVC(class_weight='balanced')

# Fit the linear support vector classifier to the training data
svc.fit(x_train, y_train)

# Print the classification report of the linear support vector classifier on the validation data
print(classification_report(y_val, svc.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Create two separate DataFrames for frauds and non-frauds using the query method
frauds = new_df.query('Class == 1')
not_frauds = new_df.query('Class == 0')

# Print the value counts of the Class column for each DataFrame
print(not_frauds['Class'].value_counts())
print(frauds['Class'].value_counts())
```

```python
# Concatenate the frauds DataFrame with a sample of non-frauds DataFrame with the same size as the frauds DataFrame
balanced_df = pd.concat([frauds, not_frauds.sample(len(frauds), random_state=1)])

# Print the value counts of the Class column in the new balanced DataFrame
print(balanced_df['Class'].value_counts())
```

```python
# Shuffle the rows of the balanced DataFrame randomly
balanced_df = balanced_df.sample(frac=1, random_state=1)

# Print the shuffled balanced DataFrame
print(balanced_df)
```

```python
# Convert the balanced DataFrame to a NumPy array
balanced_df_np = balanced_df.to_numpy()

# Split the data into training, testing, and validation sets
x_train_b, y_train_b = balanced_df_np[:700, :-1], balanced_df_np[:700, -1].astype(int)
x_test_b, y_test_b = balanced_df_np[700:842, :-1], balanced_df_np[700:842, -1].astype(int)
x_val_b, y_val_b = balanced_df_np[842:, :-1], balanced_df_np[842:, -1].astype(int)

# Print the shapes of the training, testing, and validation sets
print(x_train_b.shape, y_train_b.shape, x_test_b.shape, y_test_b.shape, x_val_b.shape, y_val_b.shape)
```

```python
# Create and fit a logistic regression model on the balanced dataset
logistic_model_b = LogisticRegression()
logistic_model_b.fit(x_train_b, y_train_b)

# Print classification report for logistic regression model on the balanced dataset
print(classification_report(y_val_b, logistic_model_b.predict(x_val_b), target_names=['Not Fraud', 'Fraud']))
```

```python
# Create and fit a random forest classifier model on the balanced dataset
rf_b = RandomForestClassifier(max_depth=2, n_jobs=-1)
rf_b.fit(x_train_b, y_train_b)

# Print classification report for random forest classifier model on the balanced dataset
print(classification_report(y_val_b, rf.predict(x_val_b), target_names=['Not Fraud', 'Fraud']))
```

```python
# import the GradientBoostingClassifier from scikit-learn
from sklearn.ensemble import GradientBoostingClassifier

# instantiate the classifier and set its hyperparameters
gbc_b = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=2, random_state=0)

# fit the model on the balanced training set
gbc_b.fit(x_train_b, y_train_b)

# make predictions on the balanced validation set and print the classification report
print(classification_report(y_val_b, gbc.predict(x_val_b), target_names=['Not Fraud', 'Fraud']))
```

```python
# instantiate a LinearSVC object with balanced class weights
svc_b = LinearSVC(class_weight='balanced')

# fit the model on the balanced training data
svc_b.fit(x_train_b, y_train_b)

# print the classification report on the balanced validation data
print(classification_report(y_val_b, svc_b.predict(x_val_b), target_names=['Not Fraud', 'Fraud']))
```
